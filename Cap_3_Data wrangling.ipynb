```python
import pandas as pd
import re
import string
import nltk
pd.set_option('display.max_colwidth', 100)

stopwords = nltk.corpus.stopwords.words('english')
ps = nltk.PorterStemmer()
import pandas as pd

fullCorpus_F = pd.read_csv('C:/Users/Mohammad/Fake.csv')
fullCorpus_T = pd.read_csv('C:/Users/Mohammad/True.csv')
fullCorpus_FC= fullCorpus_F.drop(['subject', 'date','title'], axis=1)
fullCorpus_TC= fullCorpus_T.drop(['subject', 'date','title'], axis=1)
fullCorpus_FC.head(5)
```

```python
fullCorpus_FC['label'] = 'False'
fullCorpus_TC['label'] = 'True'
fullCorpus_FC.head()
```

```python
result = fullCorpus_FC.append(fullCorpus_TC)
result.head()
```

```python
def clean_text(text):
    text = "".join([word.lower() for word in text if word not in string.punctuation])
    tokens = re.split('\W+', text)
    text = " ".join([ps.stem(word) for word in tokens if word not in stopwords])
    return text

result['cleaned_text'] = result['text'].apply(lambda x: clean_text(x))
result.head()
```

```python
from sklearn.feature_extraction.text import CountVectorizer

ngram_vect = CountVectorizer(ngram_range=(1,1))
X_counts = ngram_vect.fit_transform(result['cleaned_text'])
print(X_counts.shape)
print(ngram_vect.get_feature_names())
```

```python
data_sample = result[0:20]

ngram_vect_sample = CountVectorizer(ngram_range=(2,2))
X_counts_sample = ngram_vect_sample.fit_transform(data_sample['cleaned_text'])
print(X_counts_sample.shape)
print(ngram_vect_sample.get_feature_names())
```
