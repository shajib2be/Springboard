```python
import pandas as pd

fullCorpus_F = pd.read_csv('C:/Users/Mohammad/Fake.csv')
fullCorpus_T = pd.read_csv('C:/Users/Mohammad/True.csv')
fullCorpus_FC= fullCorpus_F.drop(['subject', 'date','title'], axis=1)
fullCorpus_TC= fullCorpus_T.drop(['subject', 'date','title'], axis=1)
fullCorpus_FC.head(5)
```

```python

fullCorpus_FC['label'] = 'False'
fullCorpus_TC['label'] = 'True'
fullCorpus_FC.head()
```

```python
result = fullCorpus_FC.append(fullCorpus_TC)
result.head()
```

```python
import string
string.punctuation
```

```python
def remove_punct(text):
    text_nopunct = "".join([char for char in text if char not in string.punctuation])
    return text_nopunct
result['body_text_clean'] = result['text'].apply(lambda x: remove_punct(x))

result.head()
```

```python
import re
import nltk
```

```python
def tokenize(text):
    tokens = re.split('\W+', text)
    return tokens

result['body_text_tokenized'] = result['body_text_clean'].apply(lambda x: tokenize(x.lower()))

result.head()
```

```python
nltk.download('stopwords')
```

```python
stopword = nltk.corpus.stopwords.words('english')

def remove_stopwords(tokenized_list):
    text = [word for word in tokenized_list if word not in stopword]
    return text

result['body_text_nostop'] = result['body_text_tokenized'].apply(lambda x: remove_stopwords(x))

result.head()
```

```python
import pandas as pd
import re
import string
ps = nltk.PorterStemmer()
pd.set_option('display.max_colwidth', 100)
def stemming(tokenized_text):
    text = [ps.stem(word) for word in tokenized_text]
    return text

result['body_text_stemmed'] = result['body_text_nostop'].apply(lambda x: stemming(x))

result.head()
```

```python
nltk.download('wordnet')
wn = nltk.WordNetLemmatizer()
def lemmatizing(tokenized_text):
    text = [wn.lemmatize(word) for word in tokenized_text]
    return text

result['body_text_lemmatized'] =result['body_text_nostop'].apply(lambda x: lemmatizing(x))

result.head(10)
```
